```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1APV3pXiAszS_0mSgkiEt9IUNH-QmyX7KwxSAwuADl6Y/gviz/tq?tqx=out:csv&sheet=medical_data')
df.head()
```

# VitD_levels

```python
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, ax=ax, stat='probability', kde=True, color= 'purple')
ax.axvline(mean, color='red', label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.set_title(f'{col}')


std = feature.std() 
plus_one_std = mean + std 
minus_one_std = mean - std

ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.legend();
```

## Questions 

1. It is continuous distribution.

2. It does not have a skew. 

3. Mesokurtic

```python
kurt_vitd = stats.kurtosis(df['VitD_levels'], fisher = False)
```

```python
kurt_vitd
```
